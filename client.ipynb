```python
import os
import socket
import time
import json
import pickle
import base64
rs = ""
with open("exp2.jpg", "rb") as imageFile:
    str = base64.b64encode(imageFile.read())
im = str.decode('utf-8')
m ={"id": 2, "image": im}
jsonObj = json.dumps(m)
msg = jsonObj.encode()
print('je suis le client numero')
soc = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
soc.connect(('localhost', 8080))
soc.sendall(msg)
time.sleep(6)
data = soc.recv(8192 * 16)
rep =data.decode('utf-8') 
jsr = json.loads(rep)
imgdata = base64.b64decode(jsr['image'])
resultImg = "response2.jpg"
with open(resultImg, 'wb') as f:
    f.write(imgdata)
print('recu : ', data)
soc.close()
```
